```python
#data.yaml      person    
#   

# train: /content/yolov5/train/images
# val: /content/yolov5/valid/images

# nc: 1
# names: ['person']
import yaml 
path = '/content/yolov5/data.yaml'
with open(path, 'r') as f : 
  data = yaml.load(f, Loader=yaml.FullLoader) 
  data['train'] = '/content/yolov5/train/images'
  data['val'] = '/content/yolov5/valid/images'
  data['nc'] = 1 
  data['names'] = ['person']

  with open(path, 'w')  as f: 
    yaml.dump(data, f)
    print(data )
```

```python
#    
from glob import glob 

train_label_list = glob('/content/yolov5/train/labels/*.txt')
for file in train_label_list:
  print(file)
  f = open(file, "rt") 
  person = []
  for line in f:
    if line.startswith("48"):
      person.append(line)
  f.close()
  
  f2 = open(file, 'wt')
  for p in person:
    f2.write(p)
  f2.close()
  f3 = open(file, "r")
  for line in f3:
    print(line)
  f3.close()
  break
```

```python
#   
from glob import glob 
import os 
train_label_list = glob('/content/yolov5/train/labels/*.txt')
for path in train_label_list:
  print(path)
  f = open(path, "rt") 
  person = []
  for line in f:
    if line.startswith("48"):
      person.append(line)
  f.close()
  if len(person) == 0 : # person    
    os.remove(path)
    os.remove(path.replace(".txt", ".jpg").replace("/labels/","/images/"))
  else : 
    f2 = open(path, 'wt')
    for p in person:
      f2.write(p.replace("48","0",1))
    f2.close()
```
